```python
import numpy as np
from numpy.fft import fft, fftfreq
import matplotlib.pyplot as plt
```

```python
#The function for generating the signal
def x(t):
    return (np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*10*t) + 0.2*np.sin(2*np.pi*20*t))
```

As the signal is a superposition of three signals with 5, 10, and 20 Hz frequencies, we expect those frequencies in the FFT spectrum.

#### Developing a Python program using Fast Fourier Transform (FFT) algorithm in NumPy package to find and plot the frequency spectrum of this signal. 

```python
#Total number of samples is 1000
#Signal duration is 1 seconds
#Sampling frequency is 1000 Hz
tmax=1
Fs=1000
ts=1/Fs
t=np.arange(0,tmax,ts)

#The signal
x_sig = x(t)
```

```python
ft=fft(x_sig)
F_sig= fftfreq(len(x_sig),d=ts)
```

```python
plt.plot(t,x_sig)
```

```python
magnitude = np.abs(ft)
positive_mask = F_sig >= 0  # Select only positive frequencies
positive_F_sig = F_sig[positive_mask]
positive_magnitude = magnitude[positive_mask]

plt.plot(positive_F_sig,positive_magnitude)
plt.xlim(0,50)
```

We found the exact same frequencies expected in part a, which are 5, 10 and 20 Hz. This is because the signal is constructed by adding three sine/cosine signals with those frequencies.

```python
#The function for generating the signal
def xm(t):
    return (1+0.5* np.sin(2*np.pi*10*t)+0.2* np.sin(2*np.pi*40*t))*np.cos(2*np.pi*100*t)
```

Carrier signal frequency (𝐹𝑐) = 100 Hz 

Modulating signal’s frequencies = 10 and 40 Hz.

List of the frequencies we can expect from the Fourier transform of the above signal: 

𝐹𝑐 = 100 Hz 

𝐹𝑐−10 = 90Hz

𝐹𝑐+10 = 110Hz

𝐹𝑐−40 = 60Hz

𝐹𝑐+40 = 140Hz

Expected bandwidth = 𝐹𝑚𝑎𝑥 – 𝐹𝑚𝑖𝑛 = 140 – 60 = 80 Hz

#### Developing a Python program using FFT algorithm in NumPy package to find and plot the frequency spectrum of this signal. 

```python
#Total number of samples is 1000
#Signal duration is 1 seconds
#Sampling frequency is 1000 Hz
tmax=1
Fs=1000
ts=1/Fs
t=np.arange(0,tmax,ts)

#The signal
x_sig = xm(t)
```

```python
plt.plot(t,x_sig)
plt.xlabel("Time")
plt.ylabel("X")
```

```python
#Fourier transforming the signal
ft=fft(x_sig)

#Finding the frequencies 
F_sig= fftfreq(len(x_sig),d=ts)
```

```python
#Need to find the magnitudes of the Fourier transformed signal
magnitude = np.abs(ft)

# Select only the positive frequencies
positive_mask = F_sig >= 0  
positive_F_sig = F_sig[positive_mask]

# Select only the magnitudes corresponding to the positive frequencies
positive_magnitude = magnitude[positive_mask]

plt.plot(positive_F_sig,positive_magnitude)
plt.xlim(0,200)
plt.xlabel("Frequencies")
plt.ylabel("Ft")
```
